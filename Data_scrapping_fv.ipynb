```
#importation of beautifulSoup 
from bs4 import BeautifulSoup
import requests
import re
import pdb
import pandas as pd
import time

Course_name = 'machine-learning'

alls = []
ratings = []


for n in range(399):
    URL = 'https://www.coursera.org/learn/' + Course_name + '/reviews?page=' + str(n)

    pages = requests.get(URL)     

    if pages.ok: 

        # Html parsing                                                         

        soup = BeautifulSoup(pages.content)                                   
        # print the resquest page with function Prettify
        #print(soup.prettify())             

        # review top part and rating top part 
            
        d = soup.body.findAll('div', attrs={'class':'top-review'} )

        parttern = re.compile(r'fill:rgba')

        for i in list(range(len(d))):
          match = parttern.findall(str(d[i]))
          star = 5 - len(match)
          if match is not None and n==1:
            ratings.append(star)

        

        for d in soup.findAll('div', attrs = {'class':'top-review'}):

          top_review = d.find('p', attrs = {'class':'top-review_comment'})
          top_review = top_review.text
          if top_review is not None and n==1:
            alls.append(top_review)
        

        # review and rating bottom part 

        d=soup.body.findAll('div',class_='_jyhj5r review review-page-review m-b-2')

        parttern=re.compile(r'fill:rgba')

        for i in list(range(len(d))):
          
          match=parttern.findall(str(d[i]))
          star = 5-len(match)
          ratings.append(star)

        for d in soup.findAll('div', attrs = {'class':'reviewText'}):
          review = d.find('p')
          review = review.text
        
        #time.sleep(3)

          if review is not None:
            alls.append(review)
          
# joint 2 columns review and rating 

data = pd.DataFrame({ Course_name : alls , 'rating': ratings})


```

```
# save data to google drive 
data.to_csv("/content/drive/MyDrive/Colab Notebooks/data.csv")
```
